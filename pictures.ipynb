```python
import anndata as ad
import matplotlib.pyplot as plt

# 读取h5ad文件
adata = ad.read_h5ad("D:\新建文件夹 (7)\m_ako_part1")

# 质荷比的值，替换为你的质荷比值
mz_value = '154.6611418882'  # 假设质荷比的值为450.3

# 确保质荷比的值在adata.var_names中
if mz_value in adata.var_names:
    # 提取质荷比的表达值
    mz_expression = adata[:, mz_value].X.toarray()[:, 0]  # 转换为稠密矩阵并取出第一列

    # 提取组织坐标
    x_coords = adata.obs['x']
    y_coords = adata.obs['y']

    # 绘制散点图
    plt.figure(figsize=(8, 6))
    plt.scatter(x_coords, y_coords, c=mz_expression, cmap='viridis', s=50)  # 使用表达值作为颜色映射

    # 添加颜色条
    plt.colorbar(label='Expression Level')

    # 添加标题和坐标轴标签
    plt.title(f'Expression of m/z {mz_value} in Tissue')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')

    # 显示图形
    plt.show()
else:
    print(f"m/z value '{mz_value}' not found in var_names.")

```
