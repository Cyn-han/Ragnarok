```python
import anndata as ad
import matplotlib.pyplot as plt

# 读取h5ad文件
adata = ad.read_h5ad("D:\\新建文件夹 (7)\\m_ako_part1")

# 获取X轴和Y轴的数据
x = adata.obs['x']
y = adata.obs['y']

# 绘制散点图
plt.figure(figsize=(10, 8))
plt.scatter(x, y, alpha=0.5)

# 添加标题和轴标签
plt.title('散点图')
plt.xlabel('X轴')
plt.ylabel('Y轴')

# 显示图形
plt.show()
```
